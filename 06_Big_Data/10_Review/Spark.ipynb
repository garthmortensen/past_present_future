1. Load the data file “senators.json”.

4. Create a new column to the table called “fullname” that will have the senators firstname and lastname together. Print the firstname, lastname, and fullname.

```
df.printSchema()
```

### Play with the Data

Count senators by party, order by count descending.

```
df.groupBy("party").count().orderBy("count", ascending=False).show()
```

List all senators with first name "John".

```
# df.select("person.firstname").show()
display(df.filter("person.firstname == 'John'"))
```

Create column "fullname" = firstname, lastname together, using [this](https://stackoverflow.com/questions/31450846/concatenate-columns-in-apache-spark-dataframe) method.   
Print the firstname, lastname, fullname.

```
from pyspark.sql import functions as sf

df = df.withColumn("fullname", 
                    sf.concat(sf.col("person.firstname"), sf.col("person.lastname")))
display(df)
```

Let's look at these name columns.

```
df.select("person.firstname", "person.lastname", "fullname").show()
```

Let's try to return senators’ first names and reversed first names.

I used [this](https://stackoverflow.com/a/59322750/5825523) solution to apply the function.

> "You have to create udf from update_email and then use it.  
> However, I'd suggest you to not use UDF fot such transformation...

```
# define the reversing function
def reverse_string(string_in):
   reversedstring=''.join(reversed(string_in))
   return reversedstring

# create a user defined function
reverse_string_udf = udf(reverse_string)

# apply the udf
df = df.withColumn("firstname_reverse", reverse_string_udf("person.firstname"))
```

```
df.select("person.firstname", "firstname_reverse").show()
```
